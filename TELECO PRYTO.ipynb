Ahora voy a eliminar las columnas que no necesito para mi posterior análisis.

```python
columnas_a_eliminar = [
    "Categoria Churn", "Razon Churn", "Estado Cliente", "Oferta", "Trimestre", "Latitud_Longitud",
    "Num Dependientes", "Num Referidos", "Metodo Pago", "Poblacion", "Estado", "Pais", "Codigo Postal",
    "Menor 30", "Factura Electronica", "Refirio Amigo", "Latitud", "Longitud"
]


df.drop(columns=columnas_a_eliminar, inplace=True)


print("Columnas restantes:")
print(df.columns)
print("\nDimensiones del DataFrame:", df.shape)
```

Observamos los valores nulos que hay. Rellenamos los valores faltantes a la columna "Tipo Internet" como 'Sin Servicio' dando por hecho que no han contratado Internet estos clientes.

```python
print(df.isnull().sum())
```

```python
df['Tipo Internet'] = df['Tipo Internet'].fillna('Sin Servicio')
```

Agrupo los servicios que el cliente contrata opcionalmente en la columna "Servicios Activos" para crear un indicador de engagement

```python
servicios = [
    "Lineas Multiples", 
    "Servicio Telefonico", 
    "Streaming Peliculas", 
    "Streaming Musica", 
    "Streaming TV",
    "Plan Proteccion Dispositivo", 
    "Backup Online", 
    "Seguridad Online", 
    "Soporte Premium", 
    "Datos Ilimitados"
]

df['Servicios'] = df[servicios].sum(axis=1)
```

Creo otra columna denominada "ARPU" = Average Revenue Per User,  que permite medir la rentabilidad mensual de cada cliente. Facilita segmentar clientes según cuánto aportan: bajo, medio o alto ARPU.

```python
df['ARPU'] = df['Ingresos Totales'] / df['Antiguedad Meses']
df['ARPU'] = df['ARPU'].fillna(0)
```

Ahora creo otra columna para segmentar a los clientes por antiguedad por baremos.

```python
bins = [0, 12, 24, 48, 60, 72, 100]
labels = ['0-12','13-24','25-48','49-60','61-72','72+']
df['Grupo_Antiguedad'] = pd.cut(df['Antiguedad Meses'], bins=bins, labels=labels, right=False)
```

Traducimos la info al español mediante mapeo.

```python
print("Valores únicos en 'Contrato':")
print(df["Contrato"].unique())

print("\nValores únicos en 'Genero':")
print(df["Genero"].unique())

print("\nValores únicos en 'Tipo Internet':")
print(df["Tipo Internet"].unique())
```

```python
map_contrato = {
    "Month-to-Month": "Mes a Mes",
    "One Year": "Un Año",
    "Two Year": "Dos Años"
}

map_genero = {
    "Female": "Mujer",
    "Male": "Hombre"
}

map_internet = {
    "DSL": "DSL",
    "Fiber Optic": "Fibra Óptica",
    "Cable": "Cable",
    "Sin Servicio": "Sin Servicio"
}

df["Contrato"] = df["Contrato"].map(map_contrato)
df["Genero"] = df["Genero"].map(map_genero)
df["Tipo Internet"] = df["Tipo Internet"].map(map_internet)
```
